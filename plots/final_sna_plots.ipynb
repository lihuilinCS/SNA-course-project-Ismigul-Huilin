```python
import matplotlib.pyplot as plt
from matplotlib import collections as mc
import numpy as np
import networkx as nx
from scipy.io import mmread
```

```python
def readCoord_dict(filename):
    corfile = open(filename)
    dic = {}
    suffix = filename.split(".")[-1]
    #print(suffix)
    if suffix == "coord" :
        for line in corfile.readlines():
            l = line.strip()
            l = l.split("\t")
            dic[l[0]] = np.array([float(l[1]), float(l[2])])
        corfile.close()
        
    if suffix == "txt" :
        for line in corfile.readlines():
            l = line.strip()
            l = l.split("\t")
            dic[int(l[2])] = np.array([float(l[0]), float(l[1])])
        corfile.close()
    return dic
```

```python
def BL_plot(coord_filename, edges_filename):
    id_xy_dic= readCoord_dict(coord_filename)
    #graph = mmread(edges_filename)
    name = edges_filename.split("/")[-1]
    
    suffix = edges_filename.split(".")[-1]
    
    if suffix == "mtx":   
        graph = mmread(edges_filename)
        lines = []
        for j,k in zip(*graph.nonzero()):
            if j>k:
                lines.append([id_xy_dic.get(j+1), id_xy_dic.get(k+1)])
                
    if suffix == "txt":
        corfile = open(edges_filename)
        lines = []
        for line in corfile.readlines():
            l = line.strip()
            l = l.split("\t")
            j = int(l[0])
            k = int(l[1])
            lines.append([id_xy_dic.get(j+1), id_xy_dic.get(k+1)])
            
    xs = [id_xy_dic.get(d)[0] for d in id_xy_dic]
    ys = [id_xy_dic.get(d)[1] for d in id_xy_dic]

            
    lc = mc.LineCollection(lines, linewidths=.3, color="blue")
    fig, ax = plt.subplots()
    
    ax.set_xlim(min(xs), max(xs))
    ax.set_ylim(min(ys), max(ys))

    ax.add_collection(lc)
    ax.autoscale()
    plt.axis('off')

    plt.savefig('plots/BL_' +  name + '.png')
    plt.show()
```

```python
def OO_plot(coord_filename, edges_filename):
    id_xy_dic= readCoord_dict(coord_filename)
    #graph = mmread(edges_filename_mtx)
    name = edges_filename.split("/")[-1]
    
    
    
    suffix = edges_filename.split(".")[-1]
    
    if suffix == "mtx":   
        graph = mmread(edges_filename)
        lines = []
        for j,k in zip(*graph.nonzero()):
            if j>k:
                source = str(j)
                target = str(k)
                if id_xy_dic.get(source) is not None and id_xy_dic.get(target) is not None:
                    lines.append([id_xy_dic.get(source), id_xy_dic.get(target)])
                
    if suffix == "txt":
        corfile = open(edges_filename)
        lines = []
        for line in corfile.readlines():
            l = line.strip()
            l = l.split("\t")
            j = l[0]
            k = l[1]
            if id_xy_dic.get(j) is not None and id_xy_dic.get(k) is not None:
                lines.append([id_xy_dic.get(j), id_xy_dic.get(k)])
    

    
    xs = [id_xy_dic.get(d)[0] for d in id_xy_dic]
    ys = [id_xy_dic.get(d)[1] for d in id_xy_dic]

            
    lc = mc.LineCollection(lines, linewidths=.3, color="blue")
    fig, ax = plt.subplots()
    
    ax.set_xlim(min(xs), max(xs))
    ax.set_ylim(min(ys), max(ys))

    ax.add_collection(lc)
    ax.autoscale()
    plt.axis('off')

    plt.savefig('plots/OO_' +  name + '.png')
    plt.show()
    
    
    return #lines
```
