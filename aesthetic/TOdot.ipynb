```python
def TXTreadCoors_dic(coors_path):
    corfile = open(coors_path)
    dic ={}
    for line in corfile.readlines():
        l = line.split("\t")
        ID = int(l[2].rstrip())-1
        dic[ID] = ([float(l[0]), float(l[1])])
    
    corfile.close()
    return dic

def COORDreadCoors_dic(coors_path):
    corfile = open(coors_path)
    dic ={}
    for line in corfile.readlines():
        l = line.split("\t")
        ID = int(l[0].rstrip())
        dic[ID] = ([float(l[1]), float(l[2])])
    
    corfile.close()
    return dic

def getG(g_path):
    graph = mmread(g_path)
    G_old = nx.Graph()
    for i in range(mminfo(g_path)[0]):
        G_old.add_node(i)
    for i,j in zip(*graph.nonzero()):
        if i > j:
            G_old.add_edge(i, j)
            
    return G_old
```

```python
def getDOT(G_old, coors_dic, J):
    
    if J =='OO':
        print('OO')
        for k in G_old.nodes():
            if coors_dic.__contains__(k):
                pos = ['x ='+ str(coors_dic[k][0])  +"," + 'y='+ str(coors_dic[k][1])]
                nx.set_node_attributes(G_old, {k:pos},'pos')
                #nx.set_node_attributes(G_old, {k:""}, "graphics")
                #nx.set_node_attributes(G_old, {k:k}, "label")
                #nx.set_node_attributes(G_old, {k:float(10)}, "width")
                #nx.set_node_attributes(G_old, {k:float(10)}, "height")
                
    if J == 'BL':
        print('BL')
        for k in G_old.nodes():

            pos = ['x ='+ str(coors_dic[k][0])  +"," + 'y='+ str(coors_dic[k][1])]
            nx.set_node_attributes(G_old, {k:pos})
            #nx.set_node_attributes(G_old, {k:""}, "graphics")
            #nx.set_node_attributes(G_old, {k:k}, "label")
            #nx.set_node_attributes(G_old, {k:float(10)}, "width")
            #nx.set_node_attributes(G_old, {k:float(10)}, "height")
        
    return nx.Graph(G_old)
        
    
```
